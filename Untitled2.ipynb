```python
# Let's load the uploaded file to check its contents and visualize the clusters.
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'synthetic_all_clusters_data.csv'
data = pd.read_csv(file_path)

# Checking the contents of the dataset
data.head()
```

```python

# Separating features and target
X = data.drop(columns=['Cluster'])
y = data['Cluster']

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('PCA Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()



cluster_labels = {
    0: "Computational dan Artificial Intelligence",
    1: "Networking dan Security",
    2: "Software Engineering dan Mobile Computing",
    3: "Information System dan Data Spasial"
}

# Map the cluster numbers to their corresponding labels
y_labeled = y.map(cluster_labels)

# Create a scatter plot with labeled clusters
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('KK Cluster Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Adding a legend with the cluster labels
handles, _ = scatter.legend_elements()
legend_labels = [cluster_labels[i] for i in range(len(cluster_labels))]
plt.legend(handles, legend_labels, title="Clusters", loc="best")

plt.grid(True)
plt.show()
```
