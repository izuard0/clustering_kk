```python
# Attempting to simplify the code to focus on essential parts and avoid errors.
import pandas as pd
from sklearn.cluster import KMeans
#import ace_tools as tools

# Load dataset
file_path = 'synthetic_training_data_with_nama_nim.csv'
data = pd.read_csv(file_path)
```

```python
# Extract relevant features for clustering (assuming non-feature columns are 'Nama' and 'NIM')
features = data.drop(columns=['Nama', 'NIM'])

# Perform KMeans clustering into 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
data['Cluster'] = kmeans.fit_predict(features)
```

```python

# Radar chart visualization for the new data point
labels = ['Cluster'] + list(data_new_features.keys())
num_vars = len(labels)

# Get values for radar chart
values = [new_cluster[0]] + data_new_df.iloc[0, :-1].values.flatten().tolist()
values += values[:1]  # Ensure the radar chart is a closed loop

# Calculate angle for each axis
theta = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
theta += theta[:1]

# Plot radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(theta, values, linewidth=2, linestyle='solid')
ax.fill(theta, values, alpha=0.25)

ax.set_xticks(theta[:-1])
ax.set_xticklabels(labels, fontsize=8)
ax.set_title(f'Radar Chart for New Data Point - Cluster {new_cluster[0]}', size=14, weight='bold', position=(0.5, 1.1),
             horizontalalignment='center', verticalalignment='center')

plt.show()
```

```python
# Misalkan 'responses' adalah dictionary yang berisi jawaban dari slider-skala 1-5
responses = {
    "q1": 3, "q2": 4, "q3": 5, "q4": 3, "q5": 2, "q6": 4, "q7": 5, "q8": 3, "q9": 4, "q10": 2,
    "q11": 4, "q12": 3, "q13": 5, "q14": 4, "q15": 3, "q16": 2, "q17": 4, "q18": 3, "q19": 5, "q20": 4
}

# Cluster labels untuk keahlian
cluster_labels = {
    0: "Computational dan Artificial Intelligence",
    1: "Networking dan Security",
    2: "Software Engineering dan Mobile Computing",
    3: "Information System dan Data Spasial"
}

# Tentukan pertanyaan mana yang relevan untuk setiap cluster
cluster_questions = {
    0: ["q1", "q3", "q9", "q20"],  # Computational dan AI
    1: ["q2", "q8", "q15"],         # Networking dan Security
    2: ["q4", "q11", "q16", "q13"], # Software Eng. dan Mobile Computing
    3: ["q10", "q12", "q7", "q19"]  # Information Systems dan Data Spasial
}

# Menghitung skor rata-rata untuk setiap cluster
cluster_scores = {}

for cluster, questions in cluster_questions.items():
    total_score = sum([responses[q] for q in questions])
    average_score = total_score / len(questions)
    cluster_scores[cluster] = average_score

# Menentukan cluster mana yang memiliki skor tertinggi
max_cluster = max(cluster_scores, key=cluster_scores.get)
chosen_cluster = cluster_labels[max_cluster]

print(f"Kelompok keahlian yang paling sesuai adalah: {chosen_cluster} dengan skor rata-rata {cluster_scores[max_cluster]:.2f}")
```
